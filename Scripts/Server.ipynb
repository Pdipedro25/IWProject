# Server

```python
import socket                   # Import socket module
import os
import pickle
HEADERSIZE = 40
port = 60000                    # Reserve a port for your service.
s = socket.socket()             # Create a socket object
host = socket.gethostname()     # Get local machine name
s.bind((host, port))            # Bind to the port
s.listen(5)                     # Now wait for client connection.

loaded_model = pickle.load(open("./classifiers/GaussianNB", 'rb'))    # Model for image classification


print ('Server listening....')

while True:
    conn, addr = s.accept()     # Establish connection with client.
    print ('Got connection from', addr)
    
    header = conn.recv(HEADERSIZE)
    
    filepath = './ReceivedImages/' + header.decode()
    
    
    with open(filepath, 'wb') as f:
        print ('file opened')
        
        print('receiving data...')
        while True:
            data = conn.recv(1024)
            
            if not data:
                break
             # write data to a file
            f.write(data)

    f.close()
    print('Successfully get the file')
    
    result = loaded_model.predict([extract_plot_hist(Image.open(filepath,'wb'))])
    if(!result):
        os.remove(filepath)
        print("File Removed!")
    else:
        print("File saved!")
    conn.close()

```
